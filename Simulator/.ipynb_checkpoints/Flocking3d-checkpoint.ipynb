```python
import numpy as np
from numpy.linalg import norm
from scipy.spatial.distance import pdist, squareform
import math
import matplotlib               
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import random
from mpl_toolkits.mplot3d import Axes3D

matplotlib.rcParams['animation.writer'] = 'ffmpeg'
plt.rcParams["animation.html"] = "html5"
```

```python
# Parameter Initialisation
timesteps = 100

num_preds = 1             #number_predators
num_bobs = 200            #flock size
bob_size = 0.25           #bird size
sense_range = 50          
sense_range_pred = 80
avoidence_range = 30
max_vel_pred = 100
max_vel = 100
min_vel = 10
close_enough = 5
dt = 0.01

neighbor_num_sense = True
sense_number = 7             #number of observed neighbors

predator_radius = 50

# rule weights flock
weight_cohesion = 0.5
weight_separation = 2
weight_alignment = 0.3
weight_random = 5
weight_prev_vel = 5
weight_predator = 8
weight_center = 0.05

#rule weights predator
weight_cohesion_pred = 0
weight_attack_boid = 0
weight_move_smooth_pred = 2
weight_prev_pred_vel = 2


#world size
x_range = 200
y_range = 200
z_range = 200

center = [100, 100, 100]

x_pos = random.sample(range(0, x_range), num_bobs)
y_pos = random.sample(range(0, y_range), num_bobs)
z_pos = random.sample(range(0, z_range), num_bobs)
bobs_pos = np.array([[i, j, k] for i, j, k in zip(x_pos, y_pos, z_pos)])          # 'bob_pos' contains the locations of all robots
bobs_vel = np.random.rand(num_bobs, 3) * max_vel                     # 'bobs_vel' contains the velocity of each robots

x_pos_pred = random.sample(range(0, x_range), num_preds)
y_pos_pred = random.sample(range(0, y_range), num_preds)
z_pos_pred = random.sample(range(0, z_range), num_preds)  
preds_pos = np.array([[i, j, k] for i, j, k in zip(x_pos_pred, y_pos_pred, z_pos_pred)])  
preds_vel = np.random.rand(num_preds, 3) * max_vel_pred
```

```python
def plot_raw():
    global ax
    ax = fig.add_subplot(111, projection='3d')
    ax.set_autoscale_on(False)
    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')
    ax.set_xlim3d([-50.0, 250.0])
    ax.set_ylim3d([-50.0, 250.0])
    ax.set_zlim3d([-50.0, 250.0])
    
def plot_update():
    fig.clf()
    plot_raw()
 
    #Plot birds  
    for index, boid in enumerate(bobs_pos):      
        bx = boid[0]
        by = boid[1]
        bz = boid[2] 
        ax.scatter(bx, by, bz, c='b', depthshade=True)
 
    #Plot predators
    if len(preds_pos) > 0:
        for index, predator in enumerate(preds_pos):     
            px = predator[0]
            py = predator[1]
            pz = predator[2] 
            ax.scatter(px, py, pz, s=50,color='red', depthshade=True)    
```

```python
#Show initial formation
fig = plt.figure(figsize = (10, 10))
plot_update()
plt.show()
```

```python
def get_center_of_flock(flock):
    flock_pos = [bobs_pos[i] for i in flock]
    centroid = np.sum(flock_pos, axis = 0) /len(flock)
    return centroid

def get_center_of_flock_vel(flock):
    flock_vel = [bobs_vel[i] for i in flock]
    centroid = np.sum(flock_vel, axis = 0) /len(flock)
    return centroid

def get_distance_matrix():
    distances = pdist(bobs_pos)
    distance_matrix = squareform(distances)
    return distance_matrix

def limitvelocity(vel):
    if np.linalg.norm(vel) > max_vel:
        vel = (vel/np.linalg.norm(vel))*max_vel
    return vel

def get_unit_vector(vector):
    return vector/norm(vector)
```

```python
def simulate(t):
    global bobs_pos
    global bobs_vel
    global preds_pos
    global preds_vel
    global distance_matrix
    
    new_bobs_pos = np.zeros((num_bobs, 3))
    new_bobs_vel = np.zeros((num_bobs, 3))
    
    new_preds_pos = np.zeros((num_preds, 3))
    new_preds_vel = np.zeros((num_preds, 3))  
    
    distance_matrix = get_distance_matrix()
    
    for i in range(len(preds_pos)):
        pred_position = preds_pos[i]
        pred_velocity = preds_vel[i]
        
        #Predator RULE 1 Cohesion: Move torwards center of flock
        #prule1 = np.zeros(3)
        #prule1 = (get_center_of_flock(list(range(num_bobs))) - pred_position) * weight_cohesion_pred

        #Predator RULE 2: Attack birds within range
        #prule2 = np.zeros(3)
        #for bob_position in bobs_pos:
        #    difference  = pred_position - bob_position
        #    distance    = np.linalg.norm(difference)
        #    if distance < sense_range_pred:
        #        prule2 = prule2 - get_unit_vector(difference)
        #prule2 = prule2 * weight_attack_boid

        #Predator RULE 3: Move the predator around the center of the world
        prule3 = np.zeros(3)
        ts = (t/float(timesteps))*4*math.pi + i*(math.pi/4.0)
        prule3[0] = (2.0+math.cos(3.0*ts))*math.cos(2.0*ts)
        prule3[1] = (2.0+math.cos(3.0*ts))*math.sin(2.0*ts)
        prule3[2] = math.sin(3*ts)
        prule3 = prule3 * weight_move_smooth_pred

        # Move the predator  
        velocity = max_vel * prule3
        #velocity = limitvelocity(velocity)
        new_preds_pos[i] = pred_position + velocity * dt
        new_preds_vel[i] = velocity

    for i in range(num_bobs):
        bob_position = bobs_pos[i]  
        bob_velocity = bobs_vel[i]
        velocity = np.zeros((3))                                 
        bob_distances = np.array(distance_matrix[i])
        close_bobs = [(index, distance) for index, distance in enumerate(bob_distances) if index != i and distance <= sense_range]
        close_bobs = sorted(close_bobs, key=lambda x: x[1])
        if neighbor_num_sense:
            local_flock = [index for k, (index, distance) in enumerate(close_bobs) if k < sense_number]
        else:
            local_flock = [index for (index, distance) in close_bobs]
        local_flock.append(i)

        #Flock RULE 1 Cohesion: Move to center of mass
        rule1 = np.zeros(3)
        rule1 = (get_center_of_flock(local_flock) - bob_position)
        rule1 = rule1 * weight_cohesion

        #Flock RULE 2 Separation: Avoid neighbors
        rule2 = np.zeros(3)
        for i2 in local_flock:
            difference  = bobs_pos[i2] - bob_position
            distance    = distance_matrix[i,i2]
            if distance < avoidence_range and i2 != i:
                rule2 = rule2 - get_unit_vector(difference)
        rule2 = rule2 * weight_separation

        #Flock RULE 3 Alignment: Steer torwards avg direction of neighbors
        rule3 = np.zeros(3)
        rule3 = (get_center_of_flock_vel(local_flock) - bob_velocity)
        rule3 = rule3 * weight_alignment

        #Flock RULE 4: Move towards the center of the grid
        rule4 = (center - bob_position) * weight_center

        #Flock RULE 5: Add randomness
        rule5 = np.random.uniform(-1,1,3)
        rule5 = rule5 * weight_random

        #Flock RULE 6: Avoid predator
        if len(preds_pos) > 0:
            rule6 = np.zeros(3)
            for pred_position in preds_pos:
                difference  = pred_position - bob_position
                distance    = np.linalg.norm(difference)
                if distance < predator_radius:
                    rule6 = (rule6 - difference)*weight_predator
        else:
            rule6 = np.zeros(3)

        #Flock RULE 7: keep previous velocity vector:
        rule7 = bob_velocity * weight_prev_vel
        
        # Move the birds
        velocity = max_vel * get_unit_vector(rule7 + rule1 + rule2 + rule3 + rule4 + rule5 + rule6)
        #velocity = limitvelocity(velocity)
        if norm(velocity) < min_vel:
            velocity = 10/get_unit_vector(velocity) * velocity
        new_bobs_pos[i] = bob_position + velocity * dt
        new_bobs_vel[i] = velocity
    
    bobs_pos = new_bobs_pos    
    bobs_vel = new_bobs_vel
    
    preds_pos = new_preds_pos    
    preds_vel = new_preds_vel
    
    return bobs_pos
```

```python
def init():
    return fig,

def update(t):
    global bobs_pos

    new_positions = simulate(t)
    plot_update()
    return (fig,)

animate = animation.FuncAnimation(fig, update, init_func=init, frames=1000, interval=8, blit=True)
```
